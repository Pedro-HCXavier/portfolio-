```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataframe.csv')
df_cortado = df.head(10)
df_cortado
```

```python
df_cortado.info()
# A coluna Revenue (usd millions) tem conteúdo em string
# É necessário transformar em número
```

```python
import numpy as np
# Criando o espaço em que o gráfico vai aparecer
figura, eixo = plt.subplots(figsize=(16,9))

nome = df_cortado.Name
# transformando essa coluna em valor numérico
receita = df_cortado['Revenue (USD millions)'].str.replace(',', '').astype(int)


eixo.barh(nome, receita, color = 'green')

# Colocando os números ao lado de cada barra
for i in eixo.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

# Estilizando o gráfico
eixo.set_title('Empresas mais ricas do mundo')

for s in ['top', 'bottom', 'left', 'right']:
    eixo.spines[s].set_visible(False)

eixo.invert_yaxis()

eixo.xaxis.set_ticks_position('none')
eixo.yaxis.set_ticks_position('none')

eixo.grid(color ='grey',
        linestyle ='-.', linewidth = 0.9,
        alpha = 0.3)

plt.xticks((100000,200000,300000,400000,500000,600000),('100M','200M','300M','400M','500M','600M'))

eixo.set_xlabel('Receita (US$)')
plt.show()
```
